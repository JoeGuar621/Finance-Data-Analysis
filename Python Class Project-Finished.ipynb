# Import Nikkei (Japanese stock exchange) Data

```python
nikkei = pd.read_csv('N225.csv')
nikkei['Date'] = pd.to_datetime(nikkei['Date'])

nikkei
```

# Clean Nikkei Data 

```python
#Convert Yen to USD
nikkei['Open'] = nikkei['Open']*0.0064
nikkei['Close'] = nikkei['Close']*0.0064
nikkei['High'] = nikkei['High']*0.0064
nikkei['Low'] = nikkei['Low']*0.0064
nikkei['Adj Close'] = nikkei['Adj Close']*0.0064


nikkei
```

# Merge Nikkei and S&P 500 Data

```python
nikkei_march = nikkei[(nikkei['Date'] >= '2020-03-01') & (nikkei['Date'] <= '2020-03-31')]


merged2 = pd.merge(df_march, nikkei_march, on='Date', how='outer', suffixes=('_spx','_nkk')).dropna()
merged2.head()
```

# Linear Regression

```python
from sklearn import linear_model
# Train the model
regression_model2 = linear_model.LinearRegression()
regression_model2.fit(X = pd.DataFrame(merged2["Close_spx"]), 
                     y = merged2["Close_nkk"])

# Check trained model y-intercept
print(regression_model2.intercept_)

# Check trained model coefficients
print(regression_model2.coef_)
```

```python
regression_model2.score(X = pd.DataFrame(merged2["Close_spx"]), 
                       y = merged2["Close_nkk"])
```

```python
#77% of variance in Nikkei explained by US Market, though coefficient is small. 
```

```python
train_prediction2 = regression_model2.predict(X = pd.DataFrame(merged2["Close_spx"]))

# Actual - prediction = residuals
residuals = merged2["Close_nkk"] - train_prediction2

residuals.describe()
```

# Conclusions

### A number of conclusions can be made about the U.S. stock market based on this analysis: 

#### 1. First, the market is resilient. Although a steep crash occurred in March of 2020, the market began to climb again (and trading volume began to increase) shortly after that. Second, although the 2020 market, in the aggregate, was valued lower than that of the 2019 market, the month-by-month growth was visually similar to the 2019 market, which may speak to the marketâ€™s ability to course correct. 

#### 2. This also speaks to the marketâ€™s susceptibility to contextual factors and changes in public policy and healthcare. The marketâ€™s periods of growth in late 2020/early-mid 2021, for example, coincided with the release of the COVID-19 vaccine and the removal of some social distancing measures and restrictions. 

#### 3. Second, the market has a strong impact on the economies of other nations. When used in a regression analysis to predict the Tokyo Stock Exchange (Nikkei), the results showed that much of the variance in Nikkei data can be explained by changes in the S&P 500. 

#### 4. However, the fact that this analysis was a univariate linear regression rather than a multivariate regression should be seen as a limitation. Before making more concrete claims, a multivariate regression should be performed, although the present analysis was limited in its data sources.
