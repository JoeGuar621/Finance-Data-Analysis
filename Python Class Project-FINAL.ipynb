```python
merged = pd.merge(df, spy, on='Date', how='outer', suffixes=('_spx','_spy'))
merged
```

```python
from sklearn import linear_model

# Train the model
regression_model = linear_model.LinearRegression()
regression_model.fit(X = pd.DataFrame(merged["Close_spx"]), 
                     y = merged["Close_spy"])

# Check trained model y-intercept
print(regression_model.intercept_)

# Check trained model coefficients
print(regression_model.coef_)
```

```python
#As SPX close increases, SPY close increased by .09
```

```python
regression_model.score(X = pd.DataFrame(merged["Close_spx"]), 
                       y = merged["Close_spy"])
```

```python
#99% of variance in spy close is explained by spx close
```

```python
#residuals
train_prediction = regression_model.predict(X = pd.DataFrame(merged["Close_spx"]))

#plot regression line
merged.plot(kind="scatter",
           x="Close_spx",
           y="Close_spy",
           figsize=(9,9),
           color="black",
           )

residuals = merged["Close_spy"] - train_prediction

residuals.describe()
# Plot regression line
plt.plot(merged["Close_spx"], 
         train_prediction, 
         color="blue");
```
