```python
import statsmodels.api as sm
from statsmodels.stats.stattools import durbin_watson
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X = df[['Inflation Rate (%)', 'Interest Rate (%)', 'GDP Growth (%)', 'Unemployment Rate (%)', 'EBITDA', 'Net Income']]
y = df['Close']
```

```python
Xc = sm.add_constant(X)

model = sm.OLS(y, Xc).fit()
print(model.summary())
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(X, y)
```

```python
def calc_residuals(model, features, label):
    predictions = model.predict(features)
    df_results = pd.DataFrame({'Actual': label, 'Predicted': predictions})
    df_results['Residuals'] = abs(df_results['Actual']) - abs(df_results['Predicted'])

    return df_results
```

```python
df_results = calc_residuals(model1, X, y)
```

```python
df_results
```

```python
def check_assumptions(results, features):

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=results['Actual'], y=results['Predicted'], color='dodgerblue', alpha=0.7)
    plt.title('Linearity of Residuals - Actual vs. Predicted')
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.show()

    sns.histplot(x=results['Residuals'], color='dodgerblue')
    plt.title('Distribution of Residuals')
    plt.show()
    

    sns.heatmap(pd.DataFrame(features).corr(), annot=True, cmap='coolwarm')
    plt.title('Multicollinearity')

    vif_data = pd.DataFrame()
    vif_data["Feature"] = features.columns
    vif_data["VIF"] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]

    print("\nVariance Inflation Factor (VIF) Table:")
    print(vif_data)

    plt.figure(figsize=(8, 6))
    plt.scatter(results.index, results['Residuals'], alpha=0.7, color='dodgerblue')
    plt.axhline(0, color='red', linestyle='--')  # Correct way to add a horizontal line
    plt.title('Homoscedasticity - Residuals vs. Index')
    plt.xlabel('Index')
    plt.ylabel('Residuals')
    plt.show()

    residuals = results['Residuals']

    dw_stat = durbin_watson(residuals)
    print(f"Test for Autocorrelation: Durbin-Watson Statistic: {dw_stat:.2f}")
```

```python
check_assumptions(df_results, X)
```
